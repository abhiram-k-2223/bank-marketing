```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/bank.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df = df.drop(columns = ['default','contact','poutcome'])
```

```python
month_mapping = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}

df['month'] = df['month'].map(month_mapping)
```

```python
df.columns
```

```python
ohe_features = ['job','marital','education','housing','loan','month']
```

```python
df = pd.get_dummies(df,columns = ohe_features,drop_first = True,dtype = np.int32)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns = ['deposit'])
y = df['deposit']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33,random_state = 42)
```

```python
x_train.corr()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter = 100000)
```

```python
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(pred,y_test)*100
score
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
lr.fit(x_train_scaled,y_train)
pred2 = lr.predict(x_test_scaled)
```

```python
score2 = accuracy_score(pred2,y_test)*100
score2
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train_scaled,y_train)
```

```python
svc_pred = svc.predict(x_test_scaled)
```

```python
svc_score = accuracy_score(svc_pred,y_test)*100
```

```python
svc_score
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train_scaled,y_train)
```

```python
knn_pred = knn.predict(x_test_scaled)
```

```python
knn_score = accuracy_score(knn_pred,y_test)*100
knn_score
```
